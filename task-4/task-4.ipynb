```python
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
lats = [20,40,-10, 35, 60]
lons = [78, -74, 120, 139, 10]
values = [10,50,30, 80, 60]


# Plot
fig = plt.figure(figsize=(10, 5))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines()
ax.add_feature(cfeature.BORDERS)

# Plot heatmap-like scatter points
sc = plt.scatter(lons, lats, c=values, cmap='hot', s=200, alpha=0.7, edgecolor='k', transform=ccrs.PlateCarree())
plt.colorbar(sc, label='Intensity')
plt.title('Geographical Heatmap using Cartopy')
plt.show()
```
